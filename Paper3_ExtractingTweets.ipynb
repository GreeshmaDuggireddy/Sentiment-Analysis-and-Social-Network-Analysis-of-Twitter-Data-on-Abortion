## Extracting Data from Twitter through SNSCRAPE

### Installing Libraries

```python
!pip install git+https://github.com/JustAnotherArchivist/snscrape.git
import os
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

### #abortion Tweets 

```python
search = '"#abortion"'

scraped_tweets = sntwitter.TwitterSearchScraper(search).get_items()

sliced_scraped_tweets = itertools.islice(scraped_tweets, 1000)

df_abortion = pd.DataFrame(sliced_scraped_tweets)
```

```python
df_abortion
```

```python
df_abortion.to_csv('df_abortion.csv', sep=',', index=False)
```

### #abortion Tweets in USA

```python
location = '40.267194,-86.134902,1000000mi'

scraped_tweets = sntwitter.TwitterSearchScraper('#abortion geocode:"{}"'.format(location)).get_items()

sliced_scraped_tweets = itertools.islice(scraped_tweets, 1000)

df_AbortionUSA = pd.DataFrame(sliced_scraped_tweets)

df_AbortionUSA['user_location'] =  df_AbortionUSA['user'].apply(lambda x: x['location'])
```

```python
df_AbortionUSA
```

```python
df_AbortionUSA.to_csv('df_AbortionUNITEDSTATES.csv', sep=',', index=False)
```
